```python
# POR SCRIPT
linear = [2, 4, 6, 8]
plt.plot(linear, '+', c='blue')

# Ambos os modos tem a mesma estrutura. matplotlib cria diversos objetos como eixos, pontos, legendas
# e formas e constroi o grafico
```

```python
linear = np.array([1, 2, 3, 4, 5])
quadratic = linear**2
plt.plot(quadratic, '-+', c = 'blue')
```

```python
# e possivel representar mais de uma coleçao de dados em um mesmo grafico
plt.plot(linear, '-+', quadratic, '-+')
```

```python
plt.plot(linear, '-+', quadratic, '-+')
# The fill betweend function
plt.gca().fill_between(range(len(linear)),linear, quadratic, facecolor = 'blue', alpha = 0.25)
# Adicionando legendas
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('test figure')
# O comando 'plt.legend('funçao 1', 'funçao 2')' adiciona uma legenda as funçoes
plt.legend('linear', 'quadratic')
# E pelos objetos do tipo "axes" que podemos manipular as legendas de eixo, por exemplo:
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
```

```python
# Observe a estrutura de fill_between() que e chamada sobre um objeto do tipo "axes":
# plt.gca() retorna um eixo sobre o qual e aplicado o metodo .fill_between()
# e os argumentos .fill_between(range(), funçao 1, funçao n, alpha = opacidade)
```
