```python
y_true=train.pop('SalePrice')
id_=train.pop('Id')
```

```python
import numpy as np
# set string values to number
def str_val_2_num(df):
    non_numerical=pd.DataFrame()
    for col in df.columns:
        if train[col].dtype == 'O':
            train[col].fillna(0)
            label=dict(zip(train[col].unique(),np.arange(len(train[col].unique()))))
            train[col]=[label[x] for x in train[col]]
            non_numerical[col]=train.pop(col)
    return non_numerical
non_num_val=str_val_2_num(train)
non_num_val
```

```python
# normalize
def normalize(df):
    for col in df.columns:
        df[col]= (df[col]-df[col].mean()) / df[col].std()
normalize(train)
train
```

```python
train[non_num_val.columns]=non_num_val[non_num_val.columns]
```

```python
train=train.fillna(0)
```

```python
train.isnull().values.any()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train.values, y_true.values, test_size=0.33, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
METRICS = [
    tf.keras.metrics.MeanAbsolutePercentageError(name='mean_absolute_percentage_error', dtype=None)
]
```

```python
#callbacks = myCallback()
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('loss')<0.0001):
      print("\nReached loss 0.0001 stop training")
      self.model.stop_training = True
        
#callback = tf.keras.callbacks.LearningRateScheduler(scheduler)
def scheduler(logs, epoch, lr):
  if logs.get('loss') > 100:
    return lr
  else:
    return lr * tf.math.exp(-0.1)

def weights_cal_and_fit(sequencial,X_train, y_train,validation_data, w_cal, w_epochs, MAX_EPOCHS):
        early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=5,mode='min')
    
        model_loss=np.zeros(w_cal)
        weights_holder=[]
        for i in range(w_cal):
            model=sequencial
            model.compile(loss='categorical_crossentropy',optimizer=tf.keras.optimizers.SGD(lr=1e-2, momentum=0.9),metrics=["accuracy"])
            temp_history = model.fit(X_train, y_train, epochs=w_epochs,validation_data=validation_data,verbose=0)
            model_loss.put(i, temp_history.history['accuracy'][-1])
            weights_holder.append(model)
        model=weights_holder[np.argmax(model_loss)]
        history = model.fit(X_train, y_train, epochs=MAX_EPOCHS,batch_size=72,validation_data=validation_data)
        return history, model

myCallback1 = myCallback()
myCallback2 = tf.keras.callbacks.LearningRateScheduler(scheduler)
```

```python
dense = keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=X_train[0].shape),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(32,kernel_regularizer=tf.keras.regularizers.L1(0.004),activity_regularizer=tf.keras.regularizers.L2(0.04)),
    tf.keras.layers.Dense(1)
])

optimizer = tf.keras.optimizers.Adam(0.01)

dense.compile(loss='mae',
            optimizer=optimizer,
            metrics=['mse'])
dense.summary()
```

```python
%%time
MAX_EPOCHS=500
history = dense.fit(X_train, y_train, epochs=MAX_EPOCHS,batch_size=32, validation_data=(X_test, y_test),verbose=2,shuffle=False, callbacks=[myCallback1, myCallback2])
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'][:], label='train')
plt.plot(history.history['val_loss'][:], label='test')
plt.legend()
plt.show()
```

```python
import matplotlib as mlt
mlt.rcParams['figure.figsize'] = (12, 10)
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
def plot_metrics(history):
  metrics =  ['loss', 'mse']
  for n, metric in enumerate(metrics):
    name = metric.replace("_"," ").capitalize()
    plt.subplot(2,2,n+1)
    plt.plot(history.epoch,  history.history[metric], color=colors[0], label='Train')
    plt.plot(history.epoch, history.history['val_'+metric],
             color=colors[0], linestyle="--", label='Val')
    plt.xlabel('Epoch')
    plt.ylabel(name)
    if metric == 'loss':
      plt.ylim([0, plt.ylim()[1]])
    
    plt.legend()
```

```python
plot_metrics(history)
```
