```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
# Use .subplot() para renderizar multiplos graficos em uma mesma imagem, os arg sao
# .subplot(linha, coluna, local_atual), observe:
plt.subplot(1, 2, 1)
# assim apartir de agora estamos em uma imagen de uma linha e duas colunas na primeira celula

some_data = np.array([1,13,23,5,4,12,11,3,2,7,8])
plt.plot(some_data)

more_data = np.sin(some_data)
# chamanos .subplot() denovo caso contrario sobrescreveremos na celula um
plt.subplot(1,2,2)
plt.plot(more_data)
```

```python
# Multiplos graficos com eixos compartilhados
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,sharex = 'True')
axs = [ax1,ax2,ax3,ax4]
```
